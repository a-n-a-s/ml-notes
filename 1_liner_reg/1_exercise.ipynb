Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
canada_percapita = pd.read_csv('canada_per_capita_income.csv')
canada_percapita
```

```python
%matplotlib inline
plt.xlabel('per capita income')
plt.ylabel('year')
plt.scatter(canada_percapita.percapitaincome,canada_percapita.year,marker="+")
```

```python
year = canada_percapita.drop('percapitaincome',axis='columns')
year 
```

```python
percapita = canada_percapita.percapitaincome
percapita
```

```python
reg = linear_model.LinearRegression()
reg.fit(year,percapita)
```

```python
reg.predict([[2020]])
```
